```
import requests
```

```
# accessing websites from url
webdata=requests.get("https://in.linkedin.com")
```

```
 webdata   # responce code of http protocol
# where 200 means url is ok
# 302 means url is redirected
# 404 means url is not found
# 5050 means server is not responding for web scrapping
```

```
# now getting actual data of any website
htmldata=webdata.text
```

```
htmldata
```

```
# import beautifulsoup (bs4)
from bs4  import BeautifulSoup
```

```
soup=BeautifulSoup(htmldata,'html5lib')   # html_data,parser
```

```
type(soup)
```

```
# finding all data
clean_data=soup.get_text()
```

```
print(clean_data)
```

```
# saving data permanently
with open('mywebdata.txt','w+') as f:
  f.write(clean_data)
```

```
# reading from a file
f=open('mywebdata.txt','r')
mydata=f.read()
f.close()
```
