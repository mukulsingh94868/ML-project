```
from urllib import request  # for downloading data from url
from bs4 import BeautifulSoup  # for souping
import time
import re
```

```
# pointing to url
url='https://en.wikipedia.org/wiki/Machine_learning'
url1='https://www.who.int/emergencies/diseases/novel-coronavirus-2019'
htmldata=request.urlopen(url1)
                        
```

```
htmldata.info
```

```
htmldata.read()  # it will download data into html format
```

```
soupdata=BeautifulSoup(htmldata,'html5lib')
# htmldata,html parser

# what is html parser -> is collection of html tags can scraps data fromparticular tag like h1,html,a,p
```

```
# now selecting a particular tag for dta scrape
atagdata=soupdata.findAll('p')
print(atagdata)
```

```
# now converting  data into string format from html format
import time
mydata=""
for i in atagdata:
  mydata += i.text
print(mydata)  
```

```
# data cleaning
clean_data=re.sub(r'\[[0-9]*\]','',mydata) # this will resolve 0 or more times number appearing in mydata
clean_data=re.sub(r'\s+','',clean_data)  # it will remove one or more white spaces with single white space
clean_data=re.sub(r'[^a-zA-Z]','',clean_data)
clean_data=re.sub(r'\s+','',clean_data)
print(clean_data)
print(type(clean_data))
```
