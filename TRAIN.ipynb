```
# extract attribute from dataframe
features=df.iloc[:,0:6].values
```

```
# now label
label=df.iloc[:,6].values
```

```
label.shape
```

```
# secratate training and testing data
from sklearn.model_selection  import train_test_split
```

```
X,x,Y,y=train_test_split(features,label,test_size=0.3)
# X is training data
# x is testing data
# Y is training answer / label
# y is testing answer / label
```

```
# calling decision tree clf
from sklearn.tree  import DecisionTreeClassifier
```

```
# calling
clf=DecisionTreeClassifier()
```

```
# now training of clf
trained=clf.fit(X,Y)
```

```
# # now we can predict
y_predict=trained.predict(x)
```
